```python
import pickle
import matplotlib.pyplot as plt

file_path = r'loss_history.p'

with open(file_path, 'rb') as f:
    loss_history = pickle.load(f)

epochs = range(1, len(loss_history) + 1)


# save epochs and loss history to a csv file
import pandas as pd
df = pd.DataFrame({'Epoch': epochs, 'Loss': loss_history})
df.to_csv('loss_history_1e-4_cnn.csv', index=False)


plt.figure(figsize=(8, 6))
plt.plot(epochs, loss_history, marker='o', linestyle='-')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss vs. Epoch")
# plt.grid(True)
plt.show()
```
